HMM:

Modelos de Markov Ocultos (HMMs) so uma classe de modelo grfico probabilstico que nos permite prever uma sequncia de variveis desconhecidas (ocultas) a partir de um conjunto de variveis observadas. Um exemplo simples de um HMM  prever o clima (varivel oculta) com base no tipo de roupa que algum usa (observado).
