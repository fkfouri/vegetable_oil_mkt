A análise de regressão diz respeito ao estudo da dependência de uma variável, **a variável dependente**, em relação a uma ou mais variáveis, **as variáveis explanatórias**, visando estimar e/ou prever o valor médio (da população) da primeira em termos dos valores conhecidos ou fixados (em amostragens repetidas) das segundas.

Mas pode ter outros nomes:


- $Y$: Variavel Dependente
- $X_k$ ou $X_i$ (corte tranversal, um ponto no tempo) ou $X_t$ (série temporal, um intervalo de tempo): Varaiaveis Exploratorias (pg 44)



## Variaveis aleatorias / estocasticas

- Variaveis aleatorias ou estocasticas: Nao pode ser totalmente explicada, independente da quantidade de variaveis explanatorias consideradas. 
- **Tem distribuição probabilística**. 
- Pode assumir qualquer valor, positivo ou negativo, dentro de um conjunto de valores com uma dada probabilidade.

- A palavra estocástica vem do grego stokhos, que significa “olho de boi”/“centro do alvo”. O resultado do lançamento de um dardo ao alvo é um processo estocástico, isto é, um processo em que nem sempre o centro do alvo é atingido.

- As variáveis aleatórias são normalmente denotadas pelas **letras maiúsculas X, Y, Z** etc., e os valores assumidos por elas são indicados pelas **letras minúsculas x, y, z** etc.

- Uma variável aleatória pode ser tanto discreta como contínua.

## correlação não é causalidade

- Causalidade: Relação entre o Evento A(causa) e Evento B (efeito)
- Correlação: fato do barulho do sino estar relacionado ao aparecimento da comida (associado pelo cachorro) não quer dizer que o primeiro cause o segundo. Na correlacao ha simetria, ou seja, nao ha distincao entre variaveis dependente e exploratorias (a aleatoriedade entre A/B ou B/A tem a mesma correlação).




[ref](https://escoladedados.org/tutoriais/correlacao-nao-e-causalidade-mas-o-que-e-entao/)


Kendall e Stuart, "uma relação estatística, por mais forte e sugestiva que seja, nunca pode estabelecer uma conexão causal: nossas ideias de causação devem vir de fora da estatística, em última análise, de alguma teoria".

## Tipo de Dados


### Série temporais:

- diarios, semanais, mensais, etc
- embora muito usado, apresentam problemas especiais para o econometrista
- a maioria dos estudos empíricos embasados nesse tipo de dado pressupõe que a série temporal subjacente seja **estacionária** _(uma série é estacionária se sua média e variância não variam sistematicamente ao longo do tempo.)_




### Corte transversal

- dados em que uma ou mais variáveis foram coletadas no mesmo ponto do tempo
- os dados em corte transversal também têm seus problemas, especificamente o da **heterogeneidade** (misturar laranja com maça).
- tem haver com dispersão
- efeito escala/magnitude

## + sobre Tipos de dados


### natureza dos dados
- experimental: o pesquisador costuma-se coleta-las mantendo-se constantes certos fatores (Ex Pressão sanguinea)
- não experimental: o pesquisador não possui controle sobre fatores (ex. PIB, taxa de desemprego, etc)

### Escalas [pg51]

- Escala de razão: Existe razão entre $X_1/X_2$ e $X_2 - X_1$. Ex.: Diferença entre PIB deste ano o anterior. Alguem ganhar 2 vezes mais que outro.

- Escala de Intervalo: 5 anos (2000 - 1995) tem sentido mas 2000/1995 nao. Uma temperatura de 90F nao é 50% maior que 60F. A escala Fahrenheit não usa 0 grau como uma base natural.

- Escala Ordinal: Tem ordenaçao mas nao pode ser quantificada. Geralmente categorica. (Alta, media e baixa)

- Escala Nominal: Não tem ordenação (Feminino, Masculino), (Casado, Solteiro)

# CAP 2 e 3 - Regressão com duas variaveis

- conhecida como bivariada ou com duas variaveis (variavels regressando/dependente se relaciona com um regressor/independente)

- $E(Y/X)$ => Valor esperado de Y dado o valor de X, ou seja, um **valor esperado condicionado** (depende dos valores de X)
- $E(Y)$ => O valor esperado ou média populacional de uma variavel aleatoria Y, **valor esperado incondicionalmente**
- $\overline{Y}$ ou $\bar{Y}$ => "Y Barra", valor médio calculado com base nos valores de uma amostra da população Y.

Exemplo. 
- Média salarial desconsiderando as faixas de renda (valor esperado incondicionalmente)
- Média salarial considerando as faixas de renda (uma média para cada faixa, valores eperados condicionalmente)

pg 58/59



## Termo Linear
- Regressão linear nos parâmetros $\beta$. São todos elevados na primeira potencia, e não é multiplicado ou dividido por nenhum outro parametro (ex $\beta_1\div\beta_2$)



## Regressao Populacional

- pressupoe que tem-se dados de toda uma população (questao empirica)
- A Funcao é composta por uma parte deterministico (sistematico) e uma parte aleatoria (não sistematico)
- A parte aleatoria tambem conhecida por **disturbio estocastico** ou **erro estocástico**.



## Regressao Amostral




- $\hat{\beta}$ => beta chapeu

- pressuposto teorico do relacionamento das variaveis
- amostra, subconjunto aleatorio de uma população

# Siglas



- FACP: Funcao de Auto-Correlação Parcial
- FEC: Funcao de esperança condicional
- FRA: Funcao de Regressa Amostra
- FRP: Funcao de Regressao Populacional

- LRP: Linha de Regressao Populacional

- MCRL: Modelo classico de regressao linear, gaussiano ou padrao (pedra angular da econometria). Variaveis X sao fixas. Modelo de regressao fixo.
- MELNT ou BLUE: melhor estimador linear não viesado (ou não tendencioso)
- MNRL: Modelo neoclassico de regressao linear (considera X aleatorio/estocastico). Modelo de regressao estocástico.
- MQO: Minimos Quadrados Ordniarios (metodo de estimaçao)
- MRL: Modelo de Regressao Linear
- MRNL: Modelo de Regressao Nao Linear
- MV: máxima verossimilhança (metodo de estimaçao)


- Navalha de Occam: principio de parcimonia, lembra um pouco o KISS (Keep it Simple). aquela que contiver o menor número possível de variáveis e hipóteses com relações lógicas entre si, das quais o fato a ser explicado segue logicamente [ref](https://pt.wikipedia.org/wiki/Navalha_de_Ockham)

- Proxy: representante de todas as variaveis omitidas ou negligenciadas que podem afetar Y. Nao foram incluidas (ou nao puderam) ser incluidas no modelo de regressao.

- RP: Regressao Populacional

- SQP: Soma dos Quadrados Previstos
- SQM: Soma dos Quadrados da Media
