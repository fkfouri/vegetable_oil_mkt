## Asia

À medida que a recuperação da pandemia de COVID-19 ganha impulso, o crescimento da renda a médio prazo na região é positivo. Combinado com a desaceleração da expansão populacional e a urbanização contínua, isso apoia a evolução contínua dos padrões alimentares, resultando no aumento da demanda por alimentos densos em calorias e nutrientes (Law, Fraser e Piracha, 2020[1]) (Kelly, 2016[2]) (Reardon et al., 2014[3]). Até 2030, a disponibilidade calórica média na região deverá aumentar em quase 200 kcal/pessoa/dia para uma média de pouco mais de 3.000 kcal, principalmente devido ao aumento no consumo de óleos vegetais, açúcar e produtos de origem animal, principalmente laticínios.

## Near East and North Africa

O crescimento de 1,7% ao ano nos próximos dez anos fará com que a população da região se aproxime de 500 milhões de pessoas até 2030. Mais da metade da população deve residir em áreas urbanas, o que pode incentivar o consumo de produtos de maior valor, incluindo carne e laticínios, mas também produtos de conveniência que contêm óleo vegetal e açúcar. A forte dependência das receitas de exportação de petróleo implica que as economias da região foram algumas das mais afetadas pela pandemia em 2020, com o PIB per capita contraindo 8%. A atividade deverá permanecer restrita em 2021 e, em média, na próxima década, crescerá apenas 1,1% a.a. Consequentemente, é improvável que constitua um grande impulsionador da demanda nos próximos dez anos.

o consumo de alimentos na América do Norte deverá permanecer em níveis elevados, aumentando em mais 48 kcal/capita/dia até 2030. ## Oilseeds and oilseed products
pg 141

O agregado de óleo vegetal neste Panorama Agrícola da OCDE-FAO inclui óleo obtido a partir do esmagamento de oleaginosas (cerca de 55% da produção mundial de óleo vegetal) e óleo de palma (36%), bem como óleos de palmiste, coco e algodão (Figura 4.1 ). Em vista de uma desaceleração na expansão da área de dendê maduro, projeta-se que o crescimento da produção na Indonésia (1,4% a.a.) e Malásia (0,9% a.a.) seja limitado. No entanto, até 2030, a Indonésia e a Malásia deverão responder por 83% da produção global de óleo de palma e 34% da produção global de óleo vegetal. Além disso, o aumento esperado na produção doméstica de biodiesel da Indonésia reduzirá o crescimento das exportações de óleo de palma bruto no médio prazo. A demanda global por óleo vegetal está projetada para expandir em 33 Mt até 2030, com o uso de alimentos representando 68% da demanda total.

De todas as commodities agrícolas, o óleo vegetal tem uma das maiores participações comerciais (41%) da produção. Indonésia e Malásia, os principais fornecedores mundiais de óleo de palma, continuarão a dominar o comércio de óleo vegetal, exportando mais de 70% de sua produção combinada e representando conjuntamente quase 60% das exportações globais. A Índia, o maior importador mundial de óleo vegetal, deverá manter seu alto crescimento de importação de 3,4% a.a. devido à crescente demanda doméstica e oportunidades limitadas de crescimento da produção.

O escopo para aumentar a produção na Indonésia e na Malásia dependerá cada vez mais das atividades de replantio de dendezeiros e das melhorias de rendimento associadas (em oposição à expansão da área). As preocupações com a sustentabilidade também influenciam a expansão da produção de óleo de palma, pois a demanda nos países desenvolvidos favorece óleos que não estão associados ao desmatamento e os consumidores buscam certificações de sustentabilidade para o óleo vegetal.



pg 43


No entanto, espera-se que o crescimento na produção de óleo de palma enfraqueça devido à crescente atenção às preocupações com a sustentabilidade e ao envelhecimento das palmeiras de óleo na Indonésia e na Malásia. Esses dois países respondem por mais de um terço da produção mundial de óleo vegetal.

No geral, a produção de óleo vegetal deverá aumentar globalmente 1,3% a.a., uma taxa mais alta do que a maioria das commodities agrícolas cobertas neste Outlook, impulsionada principalmente pela demanda de alimentos nos países em desenvolvimento resultante do crescimento da população e da renda

## Consumo 
pg 146

O consumo per capita de óleo vegetal para alimentação está projetado para crescer 0,8% a.a., consideravelmente inferior aos 2,3% a.a. aumento observado durante 2011-20 devido à demanda de alimentos cada vez mais saturada em países desenvolvidos e mercados emergentes. Na China (29 kg/capita) e no Brasil (26 kg/capita), o nível per capita de disponibilidade de alimentos com óleos vegetais deverá atingir níveis comparáveis ​​aos dos países desenvolvidos, onde o crescimento do consumo de alimentos com óleos vegetais deverá se estabilizar a 28 kg/capita, crescendo a 0,3% a.a. (Figura 4.5). 

A Índia, segundo maior consumidor mundial e primeiro importador de óleo vegetal, deverá manter um alto crescimento de consumo per capita de 2,6% a.a., atingindo 14 kg/capita até 2030. Esse aumento substancial será resultado tanto do aumento de sua produção nacional, aumento do esmagamento da produção nacional de oleaginosas e nas importações de óleo de palma principalmente da Indonésia e da Malásia. À medida que a urbanização aumenta nos países em desenvolvimento, espera-se que os hábitos alimentares e os padrões tradicionais de refeições mudem para alimentos processados com alto teor de óleo vegetal. Para os países menos desenvolvidos (PMDs), a disponibilidade per capita de óleo vegetal está projetada para aumentar em 1,3% a.a., para atingir 9 kg per capita até 2030 devido à baixa renda per capita. 


A absorção de óleo vegetal como matéria-prima para biodiesel (cerca de 10-15% do uso global de óleo vegetal) deverá permanecer estável nos próximos dez anos, em comparação com os 6,5% a.a. aumento registrado na década anterior, quando as políticas de apoio aos biocombustíveis entraram em vigor (Figura 4.6). Os aumentos projetados na Ásia e na América Latina serão compensados por reduções na Europa e na América do Norte, onde as metas fixas de mistura e o consumo de combustível de transporte em declínio afetam a demanda por biodiesel. Em geral, espera-se que as metas nacionais de consumo obrigatório de biodiesel aumentem menos do que em anos anteriores. Além disso, óleos usados, sebo e outras matérias-primas estão aumentando sua participação na produção de biodiesel, especialmente na União Européia e nos Estados Unidos, em grande parte devido a políticas específicas (consulte o Capítulo 9 para obter mais detalhes sobre biocombustíveis). A absorção de óleo vegetal pela indústria de biodiesel orientada para a exportação da Argentina está projetada para ser de 2,1 Mt até 2030, equivalente a 66% do consumo doméstico de óleo vegetal. Na Indonésia, projeta-se que o crescimento no uso de óleo vegetal para produzir biodiesel permaneça forte e atinja 7,9 Mt até 2030 devido a políticas domésticas de apoio. 


A Indonésia é o principal impulsionador do uso crescente de óleo vegetal como matéria-prima para o biodiesel no mundo. O uso de óleo vegetal como matéria-prima para biodiesel depende da definição da política (Capítulo 9) e da evolução dos preços relativos do óleo vegetal e do petróleo bruto (veja abaixo).

A Índia deverá continuar seu forte crescimento nas importações de 3,4% a.a., atingindo 21 Mt até 2030, ou cerca de um quarto das importações mundiais de óleo vegetal, a fim de responder a uma demanda crescente impulsionada pelo crescimento populacional, urbanização e aumento da renda

pg150

A pandemia de COVID-19 resultou na redução da mobilidade individual que teve fortes implicações para o consumo fora de casa. Isso pode afetar a demanda por óleo vegetal, que é amplamente utilizado para fritar. Além disso, a desaceleração temporária da atividade econômica combinada com a redução dos preços do petróleo bruto freou a demanda por óleo vegetal como matéria-prima de biodiesel. A maior parte da produção e processamento de oleaginosas e produtos é altamente mecanizada e a mobilidade da mão de obra é de menor importância. No entanto, foram relatadas interrupções importantes na colheita de óleo de palma e coco devido a restrições à mobilidade da mão de obra. As implicações de longo prazo dependerão da velocidade da recuperação econômica, pois o consumo de óleo vegetal per capita cresce fortemente com o crescimento econômico e a farinha proteica é usada como ração na produção animal mais elástica.

As políticas de biocombustíveis nos Estados Unidos, União Européia e Indonésia continuam sendo uma importante fonte de incerteza no setor de óleo vegetal, uma vez que cerca de 14% do suprimento global de óleo vegetal é destinado à produção de biodiesel. Na Indonésia, a viabilidade do mandato de 30% de biodiesel recentemente proposto é questionável, pois – além de exigir subsídios governamentais – pode impor restrições de oferta a médio prazo. Na União Européia, as reformas de políticas e o surgimento de tecnologias de biocombustíveis de segunda geração provavelmente levarão a um afastamento das matérias-primas baseadas em culturas. A evolução dos preços do petróleo bruto, que afeta a rentabilidade da produção de biodiesel, continua a ser uma importante fonte de incerteza. O crescimento mais rápido na produção de biodiesel é esperado na Indonésia, mas a relação entre os preços do óleo de palma e do petróleo bruto, bem como o desenvolvimento econômico, podem alterar consideravelmente a trajetória de crescimento projetada.
