## Oilseeds and oilseed products
pg 141

O agregado de leo vegetal neste Panorama Agrcola da OCDE-FAO inclui leo obtido a partir do esmagamento de oleaginosas (cerca de 55% da produo mundial de leo vegetal) e leo de palma (36%), bem como leos de palmiste, coco e algodo (Figura 4.1 ). Em vista de uma desacelerao na expanso da rea de dend maduro, projeta-se que o crescimento da produo na Indonsia (1,4% a.a.) e Malsia (0,9% a.a.) seja limitado. No entanto, at 2030, a Indonsia e a Malsia devero responder por 83% da produo global de leo de palma e 34% da produo global de leo vegetal. Alm disso, o aumento esperado na produo domstica de biodiesel da Indonsia reduzir o crescimento das exportaes de leo de palma bruto no mdio prazo. A demanda global por leo vegetal est projetada para expandir em 33 Mt at 2030, com o uso de alimentos representando 68% da demanda total.

De todas as commodities agrcolas, o leo vegetal tem uma das maiores participaes comerciais (41%) da produo. Indonsia e Malsia, os principais fornecedores mundiais de leo de palma, continuaro a dominar o comrcio de leo vegetal, exportando mais de 70% de sua produo combinada e representando conjuntamente quase 60% das exportaes globais. A ndia, o maior importador mundial de leo vegetal, dever manter seu alto crescimento de importao de 3,4% a.a. devido  crescente demanda domstica e oportunidades limitadas de crescimento da produo.

O escopo para aumentar a produo na Indonsia e na Malsia depender cada vez mais das atividades de replantio de dendezeiros e das melhorias de rendimento associadas (em oposio  expanso da rea). As preocupaes com a sustentabilidade tambm influenciam a expanso da produo de leo de palma, pois a demanda nos pases desenvolvidos favorece leos que no esto associados ao desmatamento e os consumidores buscam certificaes de sustentabilidade para o leo vegetal.

## Consumo 
pg 146

O consumo per capita de leo vegetal para alimentao est projetado para crescer 0,8% a.a., consideravelmente inferior aos 2,3% a.a. aumento observado durante 2011-20 devido  demanda de alimentos cada vez mais saturada em pases desenvolvidos e mercados emergentes. Na China (29 kg/capita) e no Brasil (26 kg/capita), o nvel per capita de disponibilidade de alimentos com leos vegetais dever atingir nveis comparveis aos dos pases desenvolvidos, onde o crescimento do consumo de alimentos com leos vegetais dever se estabilizar a 28 kg/capita, crescendo a 0,3% a.a. (Figura 4.5). 

A ndia, segundo maior consumidor mundial e primeiro importador de leo vegetal, dever manter um alto crescimento de consumo per capita de 2,6% a.a., atingindo 14 kg/capita at 2030. Esse aumento substancial ser resultado tanto do aumento de sua produo nacional, aumento do esmagamento da produo nacional de oleaginosas e nas importaes de leo de palma principalmente da Indonsia e da Malsia.  medida que a urbanizao aumenta nos pases em desenvolvimento, espera-se que os hbitos alimentares e os padres tradicionais de refeies mudem para alimentos processados com alto teor de leo vegetal. Para os pases menos desenvolvidos (PMDs), a disponibilidade per capita de leo vegetal est projetada para aumentar em 1,3% a.a., para atingir 9 kg per capita at 2030 devido  baixa renda per capita. 


A absoro de leo vegetal como matria-prima para biodiesel (cerca de 10-15% do uso global de leo vegetal) dever permanecer estvel nos prximos dez anos, em comparao com os 6,5% a.a. aumento registrado na dcada anterior, quando as polticas de apoio aos biocombustveis entraram em vigor (Figura 4.6). Os aumentos projetados na sia e na Amrica Latina sero compensados por redues na Europa e na Amrica do Norte, onde as metas fixas de mistura e o consumo de combustvel de transporte em declnio afetam a demanda por biodiesel. Em geral, espera-se que as metas nacionais de consumo obrigatrio de biodiesel aumentem menos do que em anos anteriores. Alm disso, leos usados, sebo e outras matrias-primas esto aumentando sua participao na produo de biodiesel, especialmente na Unio Europia e nos Estados Unidos, em grande parte devido a polticas especficas (consulte o Captulo 9 para obter mais detalhes sobre biocombustveis). A absoro de leo vegetal pela indstria de biodiesel orientada para a exportao da Argentina est projetada para ser de 2,1 Mt at 2030, equivalente a 66% do consumo domstico de leo vegetal. Na Indonsia, projeta-se que o crescimento no uso de leo vegetal para produzir biodiesel permanea forte e atinja 7,9 Mt at 2030 devido a polticas domsticas de apoio. 


A Indonsia  o principal impulsionador do uso crescente de leo vegetal como matria-prima para o biodiesel no mundo. O uso de leo vegetal como matria-prima para biodiesel depende da definio da poltica (Captulo 9) e da evoluo dos preos relativos do leo vegetal e do petrleo bruto (veja abaixo).

A ndia dever continuar seu forte crescimento nas importaes de 3,4% a.a., atingindo 21 Mt at 2030, ou cerca de um quarto das importaes mundiais de leo vegetal, a fim de responder a uma demanda crescente impulsionada pelo crescimento populacional, urbanizao e aumento da renda

A pandemia de COVID-19 resultou na reduo da mobilidade individual que teve fortes implicaes para o consumo fora de casa. Isso pode afetar a demanda por leo vegetal, que  amplamente utilizado para fritar. Alm disso, a desacelerao temporria da atividade econmica combinada com a reduo dos preos do petrleo bruto freou a demanda por leo vegetal como matria-prima de biodiesel. A maior parte da produo e processamento de oleaginosas e produtos  altamente mecanizada e a mobilidade da mo de obra  de menor importncia. No entanto, foram relatadas interrupes importantes na colheita de leo de palma e coco devido a restries  mobilidade da mo de obra. As implicaes de longo prazo dependero da velocidade da recuperao econmica, pois o consumo de leo vegetal per capita cresce fortemente com o crescimento econmico e a farinha proteica  usada como rao na produo animal mais elstica.

As polticas de biocombustveis nos Estados Unidos, Unio Europia e Indonsia continuam sendo uma importante fonte de incerteza no setor de leo vegetal, uma vez que cerca de 14% do suprimento global de leo vegetal  destinado  produo de biodiesel. Na Indonsia, a viabilidade do mandato de 30% de biodiesel recentemente proposto  questionvel, pois  alm de exigir subsdios governamentais  pode impor restries de oferta a mdio prazo. Na Unio Europia, as reformas de polticas e o surgimento de tecnologias de biocombustveis de segunda gerao provavelmente levaro a um afastamento das matrias-primas baseadas em culturas. A evoluo dos preos do petrleo bruto, que afeta a rentabilidade da produo de biodiesel, continua a ser uma importante fonte de incerteza. O crescimento mais rpido na produo de biodiesel  esperado na Indonsia, mas a relao entre os preos do leo de palma e do petrleo bruto, bem como o desenvolvimento econmico, podem alterar consideravelmente a trajetria de crescimento projetada.
