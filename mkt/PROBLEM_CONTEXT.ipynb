## Preços do óleo vegetal disparam
"A guerra em curso entre a Rússia e a Ucrânia apoiou o óleo de palma como nunca antes, enquanto a exportação de óleo de palma da Malásia de 1 a 25 de fevereiro está aumentando para cerca de 1,1 milhão de toneladas", Anil Kumar Bagani, chefe de pesquisa da Mumbai. corretora de óleo vegetal Sunvin Group disse.

Com o principal comprador da Índia recorrendo ao óleo de palma e soja para garantir o fornecimento de óleo de cozinha antes do mês de jejum muçulmano do Ramadã, os futuros de óleo de palma do mês à vista na bolsa de commodities da Malásia subiram acima de MR 8.000/mt (US$ 1.908,40) em 1º de março, subindo 7% durante comércio da manhã 1 de março.

O contrato de maio mais negociado na bolsa Bursa Malaysia Derivatives subiu 5% a MR 6.620/mt até o meio-dia de 1º de março e estava sendo negociado bem acima de sua última alta histórica de MR 6.470/mt, alcançada em 25 de fevereiro.

Em 28 de fevereiro, o óleo de girassol FOB Mar Negro Ucrânia foi avaliado em US$ 1.950,50/mt, US$ 470,50 acima dos US$ 1.480/mt em 23 de fevereiro, mostraram dados da Platts. A Platts havia pausado as avaliações do SFO do Mar Negro a partir de 24 de fevereiro.

Este é o mais alto que o SFO FOB Black Sea atingiu desde que a Platts começou a avaliar o petróleo em 10 de dezembro de 2018.

## Compradores de óleo vegetal lutam para cobrir a interrupção do óleo de girassol

[ref](https://www.spglobal.com/commodity-insights/en/market-insights/latest-news/agriculture/031122-analysis-black-sea-conflict-may-alter-global-trade-patterns-for-grains-veg-oil?utm_campaign=oktopost-2022+russia%2fukraine+tensions&utm_content=oktopost-linkedin&utm_medium=social&utm_source=linkedin)


Cerca de 1,2 milhão de toneladas/mês de óleo de girassol está programado para ser exportado para fora do Mar Negro em março e abril, disse Thomas Mielke, chefe da Oil World, com sede em Hamburgo, em uma conferência em 9 de março.

A região normalmente embarca 10,5 milhões de toneladas/ano de óleo de girassol .

"No momento, o déficit de exportação de óleo vegetal no Mar Negro representa 13% a 14% das exportações mundiais e não pode ser substituído por outras origens sem uma redução considerável nos mandatos de biodiesel", disse Mielke.

Embora o óleo de girassol normalmente represente apenas cerca de 9% do consumo global de óleo vegetal, é uma importante fonte de óleo vegetal para a UE, bem como para muitos países do Oriente Médio e Norte da África, incluindo Turquia, Irã e Egito.

A Índia, maior compradora de óleo vegetal, aumentou as importações de óleo de girassol nos últimos três meses, uma vez que os preços do óleo de girassol caíram intermitentemente abaixo dos óleos vegetais rivais de soja e palma.

"Há um mês, esperávamos que as exportações de óleo de girassol do Mar Negro crescessem bem mais de 2 milhões de toneladas para 13,5 milhões no MY 2021-22. Agora a questão é quanto será perdido devido à invasão", disse o analista James Fry, presidente de agronegócios. consultoria LMC International disse.

"Esta é uma tempestade perfeita. Os estoques e os suprimentos de exportação de petróleo estão em baixa. O armário está vazio; portanto, não há alternativa a deixar os preços altos fazerem o trabalho de racionar a demanda para equilibrar o mercado."

## Invasão russa interrompe a indústria de óleo de girassol ucraniana

[ref](https://www.spglobal.com/commodity-insights/en/market-insights/videos/market-movers-americas/220314-american-commodities-trend-bullish-on-ukraine-conflict)

A invasão representa uma ameaça de ruptura em várias frentes para os agricultores. Com as tropas russas já tendo entrado nas principais regiões produtoras de sementes de Kharkiv e Luhansk, as sementes armazenadas nessas regiões não estarão mais acessíveis aos trituradores de exportação. Além disso, as restrições logísticas ameaçam dificultar o acesso dos agricultores a fertilizantes e outros insumos, à medida que se aproximam da época crucial de plantio.

As sementes de girassol são semeadas em abril e maio antes do verão, e a colheita geralmente começa em setembro. "Há risco de oferta e demanda para a próxima campanha quando houver algumas tensões e ações militares em áreas agrícolas - muitas estradas estão bloqueadas, empresas comerciais não estão funcionando, agricultores não podem fazer o plantio e podem reduzir a área média se algumas áreas forem ocupadas, ", disse um participante do mercado.

As principais regiões de cultivo de sementes de girassol na Ucrânia estão nas zonas de estepe e estepe florestal da Ucrânia, localizadas nas províncias centrais e orientais do país. De acordo com dados do Latifúndio e do Departamento Ucraniano de Desenvolvimento Agrícola, a Ucrânia semeou 6,5 milhões de hectares na campanha de comercialização de 2020-21 (setembro a agosto) com sementes de girassol. As principais regiões produtoras são Dnipropetrovsk, Kirovohrad, Kharkiv, Zaporizhia, Nikolaev, Luhansk, Odessa e Poltava, que juntas respondem por 62% da área total semeada do país.

Os principais portos do Mar Negro para óleo de girassol na Ucrânia são Nikolaev, Chornomorsk, Dnipro, Odessa, Kherson e Yuzhniy, localizados no sudoeste do país, na costa norte do Mar Negro.

O consumo per capita de óleo vegetal para alimentação está projetado para crescer 0,8% a.a., consideravelmente inferior aos 2,3% a.a. aumento observado durante 2011-20 devido à demanda de alimentos cada vez mais saturada em países desenvolvidos e mercados emergentes. Na China (29 kg/capita) e no Brasil (26 kg/capita), o nível per capita de disponibilidade de alimentos com óleos vegetais deverá atingir níveis comparáveis ​​aos dos países desenvolvidos, onde o crescimento do consumo de alimentos com óleos vegetais deverá se estabilizar a 28 kg/capita, crescendo 0,3% aa (Figura 4.5).

A Índia, segundo maior consumidor mundial e primeiro importador de óleo vegetal, deverá manter um alto crescimento de consumo per capita de 2,6% aa, atingindo 14 kg/capita até 2030. Esse aumento substancial será resultado tanto do aumento de sua produção nacional, esmagamento do aumento da produção nacional de oleaginosas e nas importações de óleo de palma principalmente da Indonésia e da Malásia. À medida que a urbanização aumenta nos países em desenvolvimento, espera-se que os hábitos alimentares e os padrões tradicionais de refeições mudem para alimentos processados com alto teor de óleo vegetal. Para os países menos desenvolvidos (PMDs), a disponibilidade per capita de óleo vegetal está projetada para aumentar em 1,3% a.a., para atingir 9 kg per capita até 2030 devido à baixa renda per capita. 

A absorção de óleo vegetal como matéria-prima para biodiesel (cerca de 10-15% do uso global de óleo vegetal) deverá permanecer estável nos próximos dez anos, em comparação com os 6,5% a.a. aumento registrado na década anterior, quando as políticas de apoio aos biocombustíveis entraram em vigor (Figura 4.6). Os aumentos projetados na Ásia e na América Latina serão compensados ​​por reduções na Europa e na América do Norte, onde as metas fixas de mistura e o consumo de combustível de transporte em declínio afetam a demanda por biodiesel. Em geral, espera-se que as metas nacionais de consumo obrigatório de biodiesel aumentem menos do que em anos anteriores. Além disso, óleos usados, sebo e outras matérias-primas estão aumentando sua participação na produção de biodiesel, especialmente na União Européia e nos Estados Unidos, em grande parte devido a políticas específicas (consulte o Capítulo 9 para obter mais detalhes sobre biocombustíveis). A absorção de óleo vegetal pela indústria de biodiesel orientada para a exportação da Argentina está projetada para ser de 2,1 Mt até 2030, equivalente a 66% do consumo doméstico de óleo vegetal. Na Indonésia, projeta-se que o crescimento no uso de óleo vegetal para produzir biodiesel permaneça forte e atinja 7,9 Mt até 2030 devido a políticas domésticas de apoio.
